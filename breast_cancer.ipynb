```python
#splitting data
X= df.iloc[:,2:31].values
Y= df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.multiclass import type_of_target

# Split the data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)  # Use transform instead of fit_transform for test set
```

```python
# Inspect y_train
print(y_train[:5])  # Print the first 5 elements
print(y_train.dtype)  # Print the data type of y_train
```

```python
y_type = type_of_target(y_train)
if y_type not in ["binary", "multiclass"]:
    # Example for binary classification
    y_train = (y_train > 0.5).astype(int)
    y_test = (y_test > 0.5).astype(int)
```

```python
def models(x_train, y_train):
    # Logistic regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(x_train, y_train)
    
    # Decision tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
    tree.fit(x_train, y_train)
    
    # Random forest
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
    forest.fit(x_train, y_train)
    
    # Print training accuracies
    print('[0] Logistic Regression Training Accuracy:', log.score(x_train, y_train))
    print('[1] Decision Tree Training Accuracy:', tree.score(x_train, y_train))
    print('[2] Random Forest Training Accuracy:', forest.score(x_train, y_train))
    
    return log, tree, forest
```

```python
model = models(x_train, y_train)
```

```python
#test data on confusion matrix:
from sklearn.metrics import confusion_matrix

# Iterate over each model in the model list
for i, mdl in enumerate(model):
    print(f'Model {i}')
    
    # Predict on the test set
    y_pred = mdl.predict(x_test)
    
    # Compute the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Extract true positives, true negatives, false positives, and false negatives
    tp = cm[0][0]
    tn = cm[1][1]
    fn = cm[1][0]
    fp = cm[0][1]
    
    # Print the confusion matrix
    print(cm)
    
    # Calculate and print the testing accuracy
    testing_accuracy = (tp + tn) / (tp + tn + fn + fp)
    print('Testing Accuracy:', testing_accuracy)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
print(classification_report(y_test,model[0].predict(x_test)))
print(accuracy_score(y_test,model[0].predict(x_test)))
```
